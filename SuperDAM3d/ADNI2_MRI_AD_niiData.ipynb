```python
import os
import shutil
import glob

source_dir = '/home/raytrack/.jupyter/Dynamic/AD/'
destination_dir = '/home/raytrack/.jupyter/Dynamic/AD/'

# 查找所有的.nii文件
for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.endswith('.nii'):
            source_file_path = os.path.join(root, file)
            destination_file_path = os.path.join(destination_dir, file)
            # 移动.nii文件
            shutil.move(source_file_path, destination_file_path)
            print(f"Moved: {source_file_path} to {destination_file_path}")

# 再次遍历CN目录
for root, dirs, files in os.walk(source_dir):
    for dir in dirs:
        dir_path = os.path.join(root, dir)
        # 检查文件夹内是否有.nii文件
        if not glob.glob(os.path.join(dir_path, '*.nii')):
            # 如果没有.nii文件，则删除整个文件夹
            shutil.rmtree(dir_path)
            print(f"Deleted folder: {dir_path}")
```

```python
import os
import glob
import numpy as np
import nibabel as nib

out_file = "/home/raytrack/.jupyter/Dynamic/AD/"
if not os.path.isdir(out_file):
    os.makedirs(out_file)

# 更新后的路径
path = "/home/raytrack/.jupyter/Dynamic/AD/"

# 直接在CN目录下搜索.nii文件
nii_files = glob.glob(path + "*.nii")

d3list = []
labellist = []

for nii_file in nii_files:
    # 由于现在所有的.nii文件都在同一个目录下，不需要再次寻找子目录
    pid = os.path.splitext(os.path.basename(nii_file))[0]
    
    img = nib.load(nii_file)
    imgdata = img.get_fdata()
    imgdata = np.squeeze(imgdata)

    # 如果有必要，根据数据的实际需求调整数据处理逻辑
    data = np.expand_dims(imgdata, axis=3)
    print(f"Processed {pid}, data shape: {data.shape}")
    
    if np.isnan(data).any():
        print(f"NaNs found in {pid}")

    d3list.append(data)
    labellist.append(1)  # 根据实际需要调整标签逻辑

    patient_dir = os.path.join(out_file, pid)
    if not os.path.isdir(patient_dir):
        os.makedirs(patient_dir)

    # 保存处理后的3D数据
    np.save(os.path.join(patient_dir, f"{pid}.npy"), data)

# 最后，保存所有患者的数据到一个NumPy数组文件
print(f"Total processed files: {len(d3list)}")
d3data = np.asarray(d3list)
np.save('/home/raytrack/.jupyter/Dynamic/ADNI2_MRI_AD_data_3d.npy', d3data)
print("Done")
```

```python
import os
import glob
import numpy as np
import nibabel as nib

# Set the base path for the project
base_path = "/home/raytrack/.jupyter/Dynamic"

# Update the paths to use the new base path
nii_path = os.path.join(base_path, "3dAD")
out_path = os.path.join(base_path, "Preprocessed", "AD")

# Check if the output directory exists, if not, create it
if not os.path.isdir(out_path):
    os.makedirs(out_path)

d3list = []
labellist = []

# Iterate over the nii files in the dataset
for nii_file in glob.glob(nii_path + "/*.nii"):
    pid = os.path.basename(nii_file).split('.')[0]
    
    img = nib.load(nii_file)
    imgdata = img.get_fdata()
    
    # Add a new axis for channels if needed
    data = np.expand_dims(imgdata, axis=3)
    
    if np.isnan(data).any():
        print(f"NaN found in file: {nii_file}")
    
    # Append the 3D data to the list
    d3list.append(data)
    labellist.append(1)  # Assuming '1' is the label for AD category
    
    # Create a directory for the patient ID if it doesn't exist
    patient_dir = os.path.join(out_path, pid)
    if not os.path.isdir(patient_dir):
        os.makedirs(patient_dir)
    
    # Save the 3D data as a .npy file
    npy_file_path = os.path.join(patient_dir, f"{pid}.npy")
    np.save(npy_file_path, data)

# Save all the patient data into one .npy file
d3data = np.asarray(d3list)
all_data_file = os.path.join(out_path, "ADNI2_MRI_AD_data_3d.npy")
np.save(all_data_file, d3data)

# Save the labels
AD_label = np.ones((len(d3list), 1))
label_file = os.path.join(out_path, "ADNI2_MRI_AD_label.npy")
np.save(label_file, AD_label)

print(f"Processed {len(d3list)} files. Data and labels saved successfully.")
```
